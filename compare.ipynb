```python
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
```

```python
cols=('py_version', 'implementation', 'function', 'time')
py2 = pd.read_csv('python2.csv', names=cols)
py3 = pd.read_csv('python3.csv', names=cols)
pypy = pd.read_csv('pypy.csv', names=cols)
df = pd.concat([py2, py3, pypy])
```

```python
pivoted = df.pivot_table(index=['function'], columns=['implementation', 'py_version'])
pivoted.plot(kind='bar', logy=True, ylim=(0, 0.001), figsize=(15, 5))
```

```python
pivoted = df[df.implementation == 'C'].pivot_table(index=['function'], columns=['py_version'])
pivoted.plot(kind='bar', figsize=(15, 5))
```

```python
pivoted = df[df.implementation == 'python'].pivot_table(index=['function'], columns=['py_version'])
pivoted.plot(kind='bar', figsize=(15, 5))
```
